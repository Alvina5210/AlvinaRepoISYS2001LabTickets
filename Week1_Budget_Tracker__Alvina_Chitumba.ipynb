## Part 2: Your First AI Prompt

Use an AI assistant (Gemini in Colab, ChatGPT, or GitHub Copilot) to create a program that:
- Asks the user for their name
- Asks for their monthly income  
- Prints a personalized welcome message

### 📝 Document Your AI Prompt Below:

```python
# TASK: Write your AI prompt as a comment below
"""
My AI Prompt:
[create a program in python that : 1. Asks the user for their name

2. Asks for their monthly income

3. Prints a personalized welcome message]
"""

# Paste the AI-generated code below:
def main():
    # 1. Ask the user for their name
    name = input("Please enter your name: ").strip()

    # 2. Ask for their monthly income
    # We use a try-except block to handle cases where the user doesn't enter a number
    try:
        income = input(f"Hi {name}, what is your monthly income? ")
        income_value = float(income)

        # 3. Print a personalized welcome message
        print("\n" + "="*30)
        print(f"Welcome, {name}!")
        print(f"Your monthly income is recorded as: ${income_value:,.2f}")
        print("Success! Your profile has been updated.")
        print("="*30)

    except ValueError:
        print("\nOops! Please enter a valid number for the income (e.g., 3500.50).")

if __name__ == "__main__":
    main()
```

## Part 3: Building the Budget Tracker

Use AI to help you write a program that:
- Allows users to input their monthly income
- Allows users to input up to 3 expenses  
- Calculates remaining budget
- Shows a warning if budget is negative

### 💡 Pro Tip: Be specific about requirements!

```python
# TASK: Write your detailed prompt for the budget tracker
"""
My Budget Tracker Prompt:
[]please also create a program in python that :
- Allows users to input their monthly income
- Allows users to input up to 3 expenses
- Calculates remaining budget
- Shows a warning if budget is negative
"""
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
# [Paste your code here]def main():
# Monthly Budget Calculator

def main():
    print("=== Monthly Budget Calculator ===")

    # Input monthly income
    income = float(input("Enter your monthly income: $"))

    # Input up to 3 expenses
    expenses = []
    for i in range(1, 4):
        entry = input(f"Enter expense #{i} (or press Enter to skip): $")
        if entry == "":
            break
        expenses.append(float(entry))

    # Calculate total expenses
    total_expenses = sum(expenses)

    # Calculate remaining budget
    remaining_budget = income - total_expenses

    # Display results
    print("\n--- Budget Summary ---")
    print(f"Total Income: ${income:.2f}")
    print(f"Total Expenses: ${total_expenses:.2f}")
    print(f"Remaining Budget: ${remaining_budget:.2f}")

    # Warning if budget is negative
    if remaining_budget < 0:
        print("⚠️ Warning: Your budget is negative! You are overspending.")

if __name__ == "__main__":
    main()
```

```python
# ENHANCED BUDGET TRACKER
# [Paste your enhanced code here]
# Monthly Budget Calculator

def main():
    print("=== Monthly Budget Calculator ===")

    # Input monthly income
    income = float(input("Enter your monthly income: $"))

    # Input up to 3 expenses
    expenses = []
    for i in range(1, 4):
        entry = input(f"Enter expense #{i} (or press Enter to skip): $")
        if entry == "":
            break
        expenses.append(float(entry))

    # Calculate totals
    total_expenses = sum(expenses)
    remaining_budget = income - total_expenses

    # Calculate percentage of income spent
    if income > 0:
        percent_spent = (total_expenses / income) * 100
    else:
        percent_spent = 0

    # Display results
    print("\n--- Budget Summary ---")
    print(f"Total Income: ${income:.2f}")
    print(f"Total Expenses: ${total_expenses:.2f}")
    print(f"Remaining Budget: ${remaining_budget:.2f}")
    print(f"Percentage of Income Spent: {percent_spent:.2f}%")

    # Warning if budget is negative
    if remaining_budget < 0:
        print("⚠️ Warning: Your budget is negative! You are overspending.")

if __name__ == "__main__":
    main()
```
